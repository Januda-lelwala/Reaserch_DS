### Method 3: Work with Multiple Datasets

````python
# Example 3: Combine multiple datasets for analysis
gdp = datasets['gross_domestic_product_for_sri_lanka']
inflation = datasets['inflation_consumer_prices_for_sri_lanka']
unemployment = datasets['unemployment_unemployment_total_pct_of_total_labor']

print("=" * 80)
print("COMBINING MULTIPLE DATASETS FOR ANALYSIS")
print("=" * 80)

# Merge datasets for correlation analysis
combined = gdp.copy()
combined.columns = ['Year', 'GDP']

# Merge inflation
combined = combined.merge(
    inflation.rename(columns={inflation.columns[1]: 'Inflation'}),
    left_on='Year', right_on=inflation.columns[0], how='inner'
)

# Merge unemployment
combined = combined.merge(
    unemployment.rename(columns={unemployment.columns[1]: 'Unemployment_pct'}),
    left_on='Year', right_on=unemployment.columns[0], how='inner'
)

print("\nCombined dataset preview:")
print(combined.head(10))
print(f"\nCombined shape: {combined.shape}")
print(f"Columns: {list(combined.columns)}")
````
